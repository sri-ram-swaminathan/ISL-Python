#### Q1
This problem involves hyperplanes in two dimensions.

(a) Sketch the hyperplane 1 + 3X1 − X2 = 0. Indicate the set of points for which 1 + 3X1 − X2 > 0, as well as the set of points for which 1 + 3X1 − X2 < 0.

(b) On the same plot, sketch the hyperplane −2 + X1 + 2X2 = 0. Indicate the set of points for which −2 + X1 + 2X2 > 0, as well as the set of points for which −2 + X1 + 2X2 < 0.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define the grid of x and y values
x = np.linspace(0, 2, 50)
y = np.linspace(0, 4, 50)

# Create all possible (x,y) pairs in the given limit
X, Y = np.meshgrid(x, y)

# Plot the points and the hyperplanes 
plt.scatter(X, Y, s=3, alpha=0.2, color='grey')
plt.plot(x, 1 + 3*x, color='black', label='1+3x1-x2=0')
plt.plot(x, 1 - 0.5*x, color='darkslategrey', label='-1+x1+2x2=0')

# Add points above and below both hyperplanes
plt.scatter(0.5, 2.6, color='red', s=5, label='above plane a')
plt.scatter(0.5, 2.4, color='blue', s=5, label='below plane a')
plt.scatter(1, 0.6, color='darkgreen', s=5, label='above plane b')
plt.scatter(1,0.4, color='rebeccapurple', s=5, label='below plane b')

# Set limits for proper hyperplane representation
plt.xlim(0, 2)
plt.ylim(0, 4)

# Add labels and title
plt.xlabel('X1')
plt.ylabel('X2')

# Add legend
plt.legend(loc='upper right')
plt.show()
```

#### Q3
Here we explore the maximal margin classifier on a toy data set.

(a) We are given n = 7 observations in p = 2 dimensions. For each observation, there is an associated class label. Sketch the observations

```python
# Make a dataframe containing sample toy observations
data = pd.DataFrame({'x': [3, 2, 4, 1, 2, 4, 4], 'y': [4, 2, 4, 4, 1, 3, 1], 
                     'c': ['r', 'r', 'r', 'r', 'b', 'b', 'b']})
x = data['x']
y = data['y']
z = data['c']

# Plot the points 
f, ax = plt.subplots(1)
ax.set_xlim(0, 4.5)
ax.set_ylim(0, 4.5)

for i, j, k in zip(x, y, z):
    ax.scatter(i, j, color=k)
ax.grid()
```

(b) Sketch the optimal separating hyperplane, and provide the equation for this hyperplane (of the form (9.1)).

>- We notice that there are 2 pairs of red and blue points that are equidistant. The maximum margin classifier will be equidistant to these points. 
>- The hyperplane must pass through (2,1.5) and (4,3.5). This results in a line with slope 1 and intercept -0.5
>- The equation of the maximum margin classifier is given by $-0.5+x_{1}-x_{2}=0$

```python
x_values = np.array(ax.get_xlim())  
y_values_1 = x_values - 0.5 
ax.plot(x_values, y_values_1, color='black', label='optimal hyperplane')
ax.legend()
f
```

(c) Describe the classification rule for the maximal margin classifier.

>- Classify to red if $-0.5+x_{1}-x_{2}>0$
>- Classify to blue if $-0.5+x_{1}-x_{2} \leq 0$ 

(d) On your sketch, indicate the margin for the maximal margin hyperplane.

```python
y_values_2 = x_values 
y_values_3 = x_values - 1

ax.plot(x_values, y_values_2, 'r--', label='upper margin')
ax.plot(x_values, y_values_3, 'b--', label='lower margin')
ax.legend()
f
```

(e) Indicate the support vectors for the maximal margin classifier.
> The support vectors are the ones that fall on the margin.

(f) Argue that a slight movement of the seventh observation would not affect the maximal margin hyperplane.
> Since the seventh observation (4,1) is not a support vector (does not lie in margin) it will not affect the maximal margin hyperplane. 

(g) Sketch a hyperplane that is not the optimal separating hyperplane, and provide the equation for this hyperplane
>- In order to do this we simply need to shift the optimal hyperplane upwards or downwards, changing the intercept but keeping the slope constant.
>- $-0.3 + x_{1} -x_{2}=0$

```python
y_values_4 = x_values -0.3 
ax.plot(x_values, y_values_4, color='grey', label='non-optimal hyperplane')
ax.legend()
f
```
