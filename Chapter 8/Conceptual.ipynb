#### Q3
Consider the Gini index, classification error, and entropy in a simple classification setting with two classes. Create a single plot that displays each of these quantities as a function of $\hat{p}_{m1}$. The x-axis should display $\hat{p}_{m1}$, ranging from 0 to 1, and the y-axis should display the value of the Gini index, classification error, and entropy.

*Hint: In a setting with two classes, $\hat{p}_{m1}$ = 1 âˆ’ $\hat{p}_{m2}$. You could make
this plot by hand, but it will be much easier to make in Python*
> The formulas for the above metrics are : (a) Classification error : 1 - max($\hat{p}_{mk}$) (b) Gini index : $\sum_{k=1}^{2} {p}_{mk}(1-p_{mk}) $ (c) Entropy : $- \sum_{k=1}^{2} p_{mk} log(p_{mk})$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define pm1 and pm2 
p1 = np.linspace(0,1,100)
p2 = 1-p1

# Define the functions we need to plot 
error = [1 - np.maximum(i,j) for i,j in zip(p1,p2)]
gini = [(i*(1-i) + j*(1-j)) for i,j in zip(p1,p2)]
entropy = [-i*(np.log(i)) -j*(np.log(j)) for i,j in zip(p1,p2)] 

# Plot the functions 
plt.plot(p1, error, label='Classification error');
plt.plot(p1, gini, label='Gini index');
plt.plot(p1, entropy, label='Entropy');
plt.legend(loc='upper right');
```
