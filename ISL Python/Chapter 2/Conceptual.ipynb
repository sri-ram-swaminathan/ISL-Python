#### Q1 
For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size n is extremely large, and the number of predictors p is small.
>  **Flexible** method will perform better, since the training error of a flexible method decreases with the number of observations (n). Moreover the error of an inflexible (linear) method will not decrease with n.

(b) The number of predictors p is extremely large, and the number of observations n is small.
> **Inflexible** method will perform better, since the number of observations (n) is small, using a flexible method will lead to over-fitting and large test error.

(c) The relationship between the predictors and response is highly non-linear.
> **Flexible** method will perform better, since inflexible method assumes a linear relationship in the data.

(d) The variance of the error terms i.e. $\sigma^{2}=Var(\epsilon)$ is extremely high. 
> **Inflexible** method will perform better, since a flexible method will result in over-fitting and loss of generalizability.

(b) Explain why each of the five curves has the shape displayed in part (a).
> - Bayes error : the minimum value of the irreducible error in the estimate of the data. It is independent of the method, and thereby the flexibility of the method. 
> - Bias : as flexibility increases, the method fits the data more closely, thereby reducing the error due to approximating the nature of the data. 
> - Variance : as flexibility increases, the method reduces generalizability, thereby increasing the variability of the method with different data sets. 
> - Training Error : reduces monotonically since a more flexible method fits the training data more closely.
> - Test Error : initially reduces with an increase in flexibility which reduces the bias of the method, but as the flexibility increases further the variance of the method increases resulting in over-fitting and hence large test error.

#### Q5
What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

The pros and cons of a highly flexible approach are :
> - (+) High training accuracy and low bias. 
> - (+) Training accuracy increases with more data points.
> - (-) High variance and chance of overfitting.
> - (-) Low interpretability.

The pros and cons of a less flexible approach are : 
> - (+) Low variance. 
> - (+) High interpretability.
> - (-) High bias and chance of underfitting.
> - (-) Low training accuracy.

A more flexible approach is preferred when the aim is prediction, there is a non-linear relationship between predictors and response variable & we have a large sample size. 

A less flexible approach is preferred when the aim is inference, there is large variance in the error terms & we have a small sample size with a large number of predictors.

> The Euclidian distance between 2 points $(x_{1},y_{1},z_{1})$ and $(x_{2},y_{2},z_{2})$ is given by $ \sqrt{(x_{2}-x_{1})^{2}+(y_{2}-y_{1})^{2}+(z_{2}-z_{1})^{2}} $

(b) What is our prediction with K = 1? Why?
> When K=1, the prediction is simply the label of the nearest point (majority class = only class). $x_{5}$ is closest and hence the prediction is **Green**.

(c) What is our prediction with K = 3? Why?
> When K=3, the points in consideration are $x_{5},x_{6},x_{2}$. The labels are Green, Red and Red respectively. The majority class and prediction is **Red**.

(d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or small? Why?
> As the value of K increases, the decision boundary becomes less flexible (linear). This is because the prediction for one point is an average of a large number of nearest neighbours. In this case we need a flexible (non-linear) boundary. Hence a **small** K value is best.
